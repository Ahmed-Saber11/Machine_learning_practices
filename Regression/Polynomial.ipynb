```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error,root_mean_squared_error, mean_squared_error, r2_score
```

```python
data = pd.read_csv('Real estate.csv')
```

# Convert Features

```python
# define target and inputs
x = data.drop('Y house price of unit area',axis = 1)
y = data['Y house price of unit area']
```

```python
polynomail_converter = PolynomialFeatures(degree = 2 ,include_bias=False)
```

```python
poly = polynomail_converter.fit_transform(x)
```

```python
# train model
# divide data to test and train
x_train,x_test,y_train,y_test = train_test_split(poly ,y, train_size=0.70, random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred.shape
```

```python
y_test.shape
```

```python
MSE = mean_absolute_error(y_test,y_pred)
```

```python
RMSE = root_mean_squared_error (y_test,y_pred)
```

```python
MSE = mean_squared_error(y_test,y_pred)
```

```python
R2 = r2_score(y_test,y_pred)
```

```python
print('Mean Absolute Error',MSE)
print('Root Mean squared Error',RMSE)
print('Mean Squared Error',MSE)
print('R2 Score',R2)
```
